```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from scipy.signal import butter, filtfilt


np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

num_subjects    = 200      # total subjects
sampling_rate   = 1000     # Hz
duration        = 10       # seconds
num_samples     = sampling_rate * duration

time = np.linspace(0, duration, num_samples)

window_seconds  = 2.0
step_seconds    = 1.0
window_size     = int(window_seconds * sampling_rate)
step_size       = int(step_seconds * sampling_rate)


def bandpass_filter_ppg(signal, fs, lowcut=0.3, highcut=12.0, order=2):
    """
    Gentler bandpass:
      - keeps cardiac band
      - removes very slow drift & very high freq noise
      - order=2 to avoid over-smoothing / ringing
    """
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)

def add_motion_artifacts(signal, fs, n_range=(1, 2), max_duration_s=1.0):
    """
    Add fewer, milder motion bursts:
      - spikes
      - step jumps
    """
    sig = signal.copy()
    n = len(sig)
    n_art = np.random.randint(n_range[0], n_range[1] + 1)

    for _ in range(n_art):
        max_len = int(max_duration_s * fs)
        seg_len = np.random.randint(int(0.2 * fs), max_len)
        start   = np.random.randint(0, max(1, n - seg_len))
        end     = start + seg_len

        mode = np.random.choice(["spike", "jump"])

        if mode == "spike":
            # smaller spikes than before
            sig[start:end] += np.random.normal(0, 0.4, seg_len)
        elif mode == "jump":
            offset = np.random.uniform(-0.3, 0.3)
            sig[start:end] += offset

    return sig

def create_windows_1d(signal, window_size, step_size):
    windows = []
    for start in range(0, len(signal) - window_size + 1, step_size):
        win = signal[start:start + window_size]
        windows.append(win)
    return windows


#PPG simulator with dicrotic notch & hydration effect

def simulate_ppg(
    hr,
    time,
    noise_level=0.01,
    base_amplitude=1.0,
    hrv_std_frac=0.02,
    notch_depth=0.15,
):
    """
    Generate PPG with:
      - Systolic + diastolic peaks
      - Dicrotic notch (negative dip)
      - Baseline drift
      - Noise
      - HRV
    """
    t = time
    ppg = np.zeros_like(t)

    mean_period = 60.0 / hr
    beat_start = 0.0

    while beat_start < t[-1] + mean_period:
        period = mean_period * (1.0 + np.random.normal(0.0, hrv_std_frac))
        if period <= 0:
            period = mean_period

        mask = (t >= beat_start) & (t < beat_start + period)
        beat_t = t[mask] - beat_start

        if beat_t.size > 0:
            # slightly wider Gaussians than before so filtering doesn't kill them
            systolic  = np.exp(-((beat_t - 0.25 * period) ** 2) / (0.0010 * period))
            diastolic = 0.3 * np.exp(-((beat_t - 0.55 * period) ** 2) / (0.01 * period))
            notch     = -notch_depth * np.exp(
                -((beat_t - 0.40 * period) ** 2) / (0.0012 * period)
            )

            pulse = base_amplitude * (systolic + diastolic + notch)
            ppg[mask] += pulse

        beat_start += period

    # Baseline drift + measurement noise
    drift = 0.03 * np.sin(2 * np.pi * 0.3 * t)
    ppg += drift + noise_level * np.random.normal(0, 0.2, len(t))

    return ppg


# Simulate subjects, add artifacts, FILTER, and build long DataFrame

labels = ['Hydrated'] * (num_subjects // 2) + \
         ['Dehydrated'] * (num_subjects - num_subjects // 2)
random.shuffle(labels)

records = []
subject_metadata = []

hydration_to_int = {'Hydrated': 0, 'Dehydrated': 1}

for subject_id, hydration_status in enumerate(labels, start=1):
    age = random.randint(18, 75)
    sex = random.choice(['Male', 'Female'])
    monk_skin_tone = random.randint(1, 10)
    workouts_per_week = random.randint(0, 7)

    # Base HR and hydration effect
    base_hr = random.gauss(70, 8)
    if hydration_status == 'Hydrated':
        heart_rate = base_hr + random.gauss(-5, 3)
    else:
        heart_rate = base_hr + random.gauss(+5, 3)
    heart_rate = max(45, min(120, heart_rate))

    # Noise & amplitude configuration
    noise_level = 0.005 + 0.003 * (monk_skin_tone / 10.0) - 0.001 * workouts_per_week
    noise_level = max(noise_level, 0.001)
    amplitude = 1.0 - 0.2 * (monk_skin_tone / 10.0)

    if hydration_status == 'Hydrated':
        noise_level *= 0.9
        amplitude *= 1.10
        notch_depth = 0.20 + np.random.normal(0, 0.02)
    else:
        noise_level *= 1.3
        amplitude *= 0.90
        notch_depth = 0.08 + np.random.normal(0, 0.02)

    notch_depth = float(max(0.02, min(0.30, notch_depth)))

    # Simulate "clean" cardiac PPG
    ppg_clean = simulate_ppg(
        hr=heart_rate,
        time=time,
        noise_level=noise_level,
        base_amplitude=amplitude,
        hrv_std_frac=0.02,
        notch_depth=notch_depth
    )

    # Add motion artifacts (raw signal)
    ppg_raw = add_motion_artifacts(ppg_clean, sampling_rate)

    # Gentler bandpass filter -> filtered signal
    ppg_filt = bandpass_filter_ppg(ppg_raw, sampling_rate,
                                   lowcut=0.3, highcut=12.0, order=2)

    # Normalize filtered signal to zero-mean and [-1, 1] *per subject*
    ppg_filt -= np.mean(ppg_filt)
    max_abs = np.max(np.abs(ppg_filt))
    if max_abs > 0:
        ppg_filt /= max_abs
    else:
        ppg_filt = np.zeros_like(ppg_filt)

    # Store time series rows (raw + filtered)
    for t_val, raw_val, filt_val in zip(time, ppg_raw, ppg_filt):
        records.append({
            'subject_id': subject_id,
            'time': t_val,
            'ppg_raw': raw_val,
            'ppg_filt': filt_val,
            'age': age,
            'sex': sex,
            'monk_skin_tone': monk_skin_tone,
            'workouts_per_week': workouts_per_week,
            'heart_rate': heart_rate,
            'hydration_status': hydration_status
        })

    subject_metadata.append({
        'subject_id': subject_id,
        'age': age,
        'sex': sex,
        'monk_skin_tone': monk_skin_tone,
        'workouts_per_week': workouts_per_week,
        'heart_rate': heart_rate,
        'hydration_status': hydration_status,
        'hydration_label': hydration_to_int[hydration_status],
        'noise_level_used': noise_level,
        'amplitude_used': amplitude,
        'notch_depth_used': notch_depth
    })

df = pd.DataFrame(records)
subject_df = pd.DataFrame(subject_metadata)

df.to_csv('simulated_ppg_timeseries_data.csv', index=False)
subject_df.to_csv('simulated_ppg_subject_data.csv', index=False)

print("Saved simulated_ppg_timeseries_data.csv and simulated_ppg_subject_data.csv")


# Subject-level split (NO leakage) + window datasets (filtered only)

train_subjects, val_subjects = train_test_split(
    subject_df['subject_id'].values,
    test_size=0.2,
    random_state=42,
    stratify=subject_df['hydration_label'].values
)

print("Num train subjects:", len(train_subjects))
print("Num val subjects:", len(val_subjects))

X_train_list, y_train_list = [], []
X_val_list,   y_val_list   = [], []

for sid in train_subjects:
    sdata = df[df['subject_id'] == sid]
    ppg_s  = sdata['ppg_filt'].values
    label  = hydration_to_int[sdata['hydration_status'].iloc[0]]
    wins   = create_windows_1d(ppg_s, window_size, step_size)
    for w in wins:
        X_train_list.append(w)
        y_train_list.append(label)

for sid in val_subjects:
    sdata = df[df['subject_id'] == sid]
    ppg_s  = sdata['ppg_filt'].values
    label  = hydration_to_int[sdata['hydration_status'].iloc[0]]
    wins   = create_windows_1d(ppg_s, window_size, step_size)
    for w in wins:
        X_val_list.append(w)
        y_val_list.append(label)

X_train = np.stack(X_train_list, axis=0)[..., np.newaxis]
y_train = np.array(y_train_list)
X_val   = np.stack(X_val_list, axis=0)[..., np.newaxis]
y_val   = np.array(y_val_list)

print("Train windows:", X_train.shape, "Val windows:", X_val.shape)
print("Train class balance:", {c: int((y_train == c).sum()) for c in np.unique(y_train)})
print("Val class balance:", {c: int((y_val == c).sum()) for c in np.unique(y_val)})


# Plot example PPG (raw vs filtered) for a few subjects

unique_subjects = subject_df['subject_id'].values
n_plot = min(4, len(unique_subjects))

fig, axs = plt.subplots(n_plot, 1, figsize=(12, 8), sharex=True)
for i, sid in enumerate(unique_subjects[:n_plot]):
    sdata = df[(df['subject_id'] == sid) & (df['time'] <= 5.0)]
    axs[i].plot(sdata['time'], sdata['ppg_raw'],  lw=0.7, label='Raw')
    axs[i].plot(sdata['time'], sdata['ppg_filt'], lw=0.7, label='Filtered')
    axs[i].set_title(
        f"Subject {sid} — HR: {sdata['heart_rate'].iloc[0]:.1f} bpm "
        f"— Status: {sdata['hydration_status'].iloc[0]}"
    )
    axs[i].set_ylabel("PPG")
    axs[i].legend(loc='upper right', fontsize=8)
axs[-1].set_xlabel("Time (s)")
plt.tight_layout()
plt.show()

# 7. CNN on FILTERED windows

input_shape = X_train.shape[1:]  # (window_size, 1)

inputs = keras.Input(shape=input_shape)
x = layers.Conv1D(32, 7, padding='same', activation='relu')(inputs)
x = layers.MaxPooling1D(2)(x)

x = layers.Conv1D(64, 5, padding='same', activation='relu')(x)
x = layers.MaxPooling1D(2)(x)

x = layers.Conv1D(128, 3, padding='same', activation='relu')(x)
x = layers.GlobalAveragePooling1D()(x)

x = layers.Dense(64, activation='relu', name='embedding_dense')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(1, activation='sigmoid', name='output')(x)

model = keras.Model(inputs=inputs, outputs=outputs)
model.summary()

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=8,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=4,
    min_lr=1e-6,
    verbose=1
)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=40,
    batch_size=64,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)


# Plot of CNN training curves

def plot_training(hist_obj):
    hist = hist_obj.history
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.plot(hist['loss'], label='train')
    plt.plot(hist['val_loss'], label='val')
    plt.title('CNN Loss (filtered PPG)')
    plt.xlabel('Epoch')
    plt.ylabel('Binary cross-entropy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(hist['accuracy'], label='train')
    plt.plot(hist['val_accuracy'], label='val')
    plt.title('CNN Accuracy (filtered PPG)')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_training(history)


#  CNN evaluation

val_probs_cnn = model.predict(X_val, verbose=0).flatten()
val_preds_cnn = (val_probs_cnn >= 0.5).astype(int)

acc_cnn_window = accuracy_score(y_val, val_preds_cnn)
print("\nCNN-only window-level validation accuracy:", acc_cnn_window)
print("\nCNN-only window-level classification report:")
print(classification_report(y_val, val_preds_cnn, digits=3))
print("CNN-only window-level confusion matrix:")
print(confusion_matrix(y_val, val_preds_cnn))


# Feature extractor + Gradient Boosting on embeddings

embedding_layer = model.get_layer('embedding_dense')
feature_extractor = keras.Model(
    inputs=model.input,
    outputs=embedding_layer.output
)

Z_train = feature_extractor.predict(X_train, verbose=0)
Z_val   = feature_extractor.predict(X_val, verbose=0)

print("Z_train shape:", Z_train.shape)
print("Z_val shape:", Z_val.shape)

gb_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('gb', GradientBoostingClassifier(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=3,
        subsample=0.8,
        random_state=42
    ))
])

gb_pipe.fit(Z_train, y_train)
y_val_pred_gb  = gb_pipe.predict(Z_val)
y_val_proba_gb = gb_pipe.predict_proba(Z_val)[:, 1]

acc_gb_window = accuracy_score(y_val, y_val_pred_gb)
print(f"\nCNN+GBM window-level validation accuracy: {acc_gb_window:.3f}")

print("\nCNN+GBM window-level classification report (0=Hydrated, 1=Dehydrated):")
print(classification_report(y_val, y_val_pred_gb, digits=3))
print("CNN+GBM window-level confusion matrix:")
print(confusion_matrix(y_val, y_val_pred_gb))

gb = gb_pipe.named_steps['gb']
importances = gb.feature_importances_
print("\nFeature importances shape:", importances.shape)

plt.figure(figsize=(10, 4))
plt.bar(range(len(importances)), importances)
plt.title("Gradient Boosting Feature Importances (CNN Embeddings)")
plt.xlabel("Embedding Feature Index")
plt.ylabel("Importance")
plt.tight_layout()
plt.show()

top_k = 10
idx_sorted = np.argsort(importances)[::-1]
print(f"\nTop {top_k} most important embedding features:")
for i in idx_sorted[:top_k]:
    print(f"  Feature {i}: importance = {importances[i]:.4f}")

# Subject-level evaluation (CNN vs CNN+GBM)

def subject_level_pred(ppg_signal, model, feature_extractor, gb_pipe,
                       fs=1000, window_seconds=2.0, step_seconds=1.0,
                       threshold=0.5):
    ppg_signal = np.asarray(ppg_signal).flatten()
    wsize = int(window_seconds * fs)
    ssize = int(step_seconds * fs)

    wins = create_windows_1d(ppg_signal, wsize, ssize)
    if not wins:
        raise ValueError("Signal too short for given window/step settings.")

    X_sub = np.stack(wins, axis=0)[..., np.newaxis]

    # CNN
    probs_cnn = model.predict(X_sub, verbose=0).flatten()
    subj_prob_cnn = float(np.mean(probs_cnn))
    subj_pred_cnn = int(subj_prob_cnn >= threshold)

    # Hybrid
    Z_sub   = feature_extractor.predict(X_sub, verbose=0)
    probs_gb = gb_pipe.predict_proba(Z_sub)[:, 1]
    subj_prob_gb = float(np.mean(probs_gb))
    subj_pred_gb = int(subj_prob_gb >= threshold)

    return subj_pred_cnn, subj_prob_cnn, subj_pred_gb, subj_prob_gb

y_true_subject = []
y_pred_cnn_subject = []
y_pred_gb_subject  = []

for sid in val_subjects:
    sdata = df[df['subject_id'] == sid]
    ppg_s = sdata['ppg_filt'].values
    true_label = hydration_to_int[sdata['hydration_status'].iloc[0]]

    pred_cnn, prob_cnn, pred_gb, prob_gb = subject_level_pred(
        ppg_s, model, feature_extractor, gb_pipe,
        fs=sampling_rate,
        window_seconds=window_seconds,
        step_seconds=step_seconds,
        threshold=0.5
    )

    y_true_subject.append(true_label)
    y_pred_cnn_subject.append(pred_cnn)
    y_pred_gb_subject.append(pred_gb)

y_true_subject = np.array(y_true_subject)
y_pred_cnn_subject = np.array(y_pred_cnn_subject)
y_pred_gb_subject  = np.array(y_pred_gb_subject)

acc_cnn_subject = accuracy_score(y_true_subject, y_pred_cnn_subject)
acc_gb_subject  = accuracy_score(y_true_subject, y_pred_gb_subject)

print("\nCNN subject-level accuracy:", acc_cnn_subject)
print("CNN subject-level classification report:")
print(classification_report(y_true_subject, y_pred_cnn_subject, digits=3))

print("\nCNN+GBM subject-level accuracy:", acc_gb_subject)
print("CNN+GBM subject-level classification report:")
print(classification_report(y_true_subject, y_pred_gb_subject, digits=3))

print("CNN subject-level confusion matrix:")
print(confusion_matrix(y_true_subject, y_pred_cnn_subject))
print("CNN+GBM subject-level confusion matrix:")
print(confusion_matrix(y_true_subject, y_pred_gb_subject))


# Example validation window + preds

import random as pyrand

idx = pyrand.randint(0, X_val.shape[0] - 1)
plt.figure(figsize=(10, 3))
plt.plot(np.arange(X_val.shape[1]) / sampling_rate, X_val[idx, :, 0])
plt.title("Example validation window (filtered PPG)")
plt.xlabel("Time (s)")
plt.ylabel("PPG (norm)")
plt.tight_layout()
plt.show()

print("\nTrue label (window):", y_val[idx], " (0=Hydrated, 1=Dehydrated)")
print("CNN prob (Dehydrated):", float(val_probs_cnn[idx]))
print("CNN+GBM prob (Dehydrated):", float(y_val_proba_gb[idx]))
print("CNN+GBM pred:", y_val_pred_gb[idx])


#  Compare accuracies in a bar chart (window vs subject, CNN vs hybrid)

labels_plot = ["CNN windows", "CNN+GBM windows", "CNN subjects", "CNN+GBM subjects"]
accs_plot   = [acc_cnn_window, acc_gb_window, acc_cnn_subject, acc_gb_subject]

plt.figure(figsize=(8, 4))
plt.bar(range(len(labels_plot)), accs_plot)
plt.xticks(range(len(labels_plot)), labels_plot, rotation=20)
plt.ylim(0, 1.0)
plt.ylabel("Accuracy")
plt.title("Performance Comparison: CNN vs CNN+GBM (filtered PPG)")
plt.tight_layout()
plt.show()
```

Here are the most critical decision points or transitions in the model pipeline:

1.  **Data Generation & Preprocessing (Raw to Filtered PPG)**:
    *   **Input**: Raw simulated PPG signal with added motion artifacts.
    *   **Transition**: Application of a bandpass filter (`bandpass_filter_ppg`) to remove noise and drift, and normalization.
    *   **Output**: Cleaned and normalized filtered PPG signal, which is the primary input for subsequent modeling.

2.  **Signal Windowing (Continuous to Segmented Data)**:
    *   **Input**: Continuous, filtered PPG signal for each subject.
    *   **Transition**: Creation of fixed-size, overlapping windows (`create_windows_1d`) from the continuous signal.
    *   **Output**: A dataset of individual PPG windows (`X_train`, `X_val`), each representing a short segment of the physiological signal, ready for CNN input.

3.  **CNN Model Training and Prediction (Feature Learning)**:
    *   **Input**: Windowed PPG data (`X_train`).
    *   **Transition**: Training a 1D Convolutional Neural Network (CNN) to learn features directly from the raw PPG windows and classify hydration status.
    *   **Output**: Trained CNN model capable of predicting hydration status at the window level, and its corresponding *embedding layer output*.

4.  **Feature Extraction for Hybrid Model (Embeddings for GBM)**:
    *   **Input**: Trained CNN model.
    *   **Transition**: Extraction of the latent features (embeddings) from a specific dense layer (`embedding_dense`) of the trained CNN.
    *   **Output**: A set of high-level, learned features (`Z_train`, `Z_val`) for each PPG window, representing a more abstract representation of the signal content.

5.  **Hybrid Model Training and Prediction (CNN+GBM Classification)**:
    *   **Input**: Extracted CNN embeddings (`Z_train`).
    *   **Transition**: Training a Gradient Boosting Machine (GBM) on these learned embeddings to perform the final classification.
    *   **Output**: Trained CNN+GBM model capable of predicting hydration status with potentially improved performance by combining the strengths of deep feature learning and traditional machine learning.

6.  **Window-level to Subject-level Aggregation (Decision Making)**:
    *   **Input**: Window-level predictions (probabilities) from either the CNN or the CNN+GBM model.
    *   **Transition**: Averaging the window-level probabilities for all windows belonging to a single subject to derive a single subject-level probability, followed by a thresholding (0.5) to get the final subject-level prediction.
    *   **Output**: Final subject-level predictions of hydration status, which is often the most clinically relevant outcome.
